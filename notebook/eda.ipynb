Here’s a detailed explanation of the given features, tailored to Bitcoin analysis, categorized into **Sentiment Indicators**, **Blockchain Metrics**, **Technical Indicators**, and **Macro/Market Metrics**:

**Sentiment Indicators**

1.	google_trends_bitcoin

Tracks the popularity of Bitcoin-related search terms over time. Higher values often indicate heightened public interest, which can lead to increased market activity or speculative price changes.

**Blockchain Metrics**

2.	active_addresses_blockchain

Number of unique Bitcoin addresses participating in transactions. This reflects the level of activity and adoption on the Bitcoin network. Higher values may suggest increased usage or network engagement.

3.	hash_rate_blockchain

Measures the total computational power used by miners on the Bitcoin network. A higher hash rate indicates greater network security and miner confidence. Sharp drops can suggest miner capitulation, potentially signaling bearish trends.

4.	miner_revenue_blockchain

Total earnings by Bitcoin miners (block rewards + transaction fees). It reflects the profitability of mining and can influence miner behavior, such as selling or holding mined BTC.

**Technical Indicators**

5.	btc_sma_14 **(14-Day Simple Moving Average)**

A lagging indicator that shows the average Bitcoin closing price over 14 days. It smooths out price data to identify overall trends and support/resistance levels.

6.	btc_ema_14 **(14-Day Exponential Moving Average)**

Similar to SMA but gives more weight to recent prices. EMA is often used in fast-moving markets like Bitcoin to detect short-term trends.

7.	btc_rsi_14 **(14-Day Relative Strength Index)**

Oscillates between 0 and 100, measuring the speed and change of price movements. RSI > 70 indicates overbought conditions (potential reversal); RSI < 30 suggests oversold conditions (potential recovery).

8.	btc_macd **(Moving Average Convergence Divergence)**

Represents the difference between a short-term EMA (usually 12 days) and a longer-term EMA (26 days). Positive values indicate bullish momentum.

9.	btc_macd_signal

The 9-day EMA of the MACD. Often used as a trigger line for buy/sell signals when crossed by the MACD.

10.	btc_macd_diff

The difference between btc_macd and btc_macd_signal. Positive values suggest upward momentum, while negative values suggest downward momentum.

11.	btc_bb_high**,** btc_bb_low**,** btc_bb_mid **(Bollinger Bands)**

•	**High**: Upper boundary indicating overbought conditions.

•	**Low**: Lower boundary indicating oversold conditions.

•	**Mid**: The SMA used to calculate the bands (usually 20 days). These bands measure price volatility.

12.	btc_bb_width

The distance between the upper and lower Bollinger Bands. Expanding width indicates increasing volatility, while contracting width suggests decreasing volatility.

13.	btc_atr_14 **(14-Day Average True Range)**

Measures Bitcoin’s volatility by calculating the average range between daily high and low prices over 14 days. Higher ATR suggests higher market volatility.

**Market Metrics**

14.	btc_trading_volume

Represents the total number of Bitcoin traded during a given period. High volume during price moves can confirm the trend; low volume may indicate potential reversals.

15.	btc_volatility_index

Quantifies Bitcoin’s historical price volatility. Higher values suggest more unpredictable price movements.

**Macro and Cross-Market Metrics**

16.	Gold_Close

Closing price of gold. Bitcoin and gold are often considered “safe-haven” assets, with their prices sometimes moving inversely to traditional markets.

17.	Oil_Close

Closing price of crude oil. While not directly related to Bitcoin, it reflects macroeconomic conditions that can influence overall market sentiment.

18.	DJI **(Dow Jones Industrial Average)**

Tracks 30 large U.S. companies. A barometer for traditional market performance; Bitcoin sometimes reacts inversely during risk-off periods.

19.	GSPC **(S&P 500)**

Index of 500 large U.S. companies. Like the DJI, its performance can influence Bitcoin through risk sentiment.

20.	IXIC **(NASDAQ Composite)**

Heavily tech-focused index. Bitcoin’s correlation with tech stocks can make this a key indicator for market sentiment.

21.	NYSE FANG+

Tracks high-growth tech stocks like Facebook, Amazon, Netflix, and Google. A proxy for speculative sentiment, which can influence Bitcoin prices.

22.	ARK Innovation ETF

Tracks disruptive tech companies. A popular ETF among risk-tolerant investors, its performance can correlate with Bitcoin’s speculative appeal.

23.	CBOE Volatility Index **(VIX)**

Measures expected stock market volatility. High VIX values indicate fear in the market, which can drive investors to or away from Bitcoin depending on sentiment.

24.	iShares MSCI Emerging Markets ETF

Tracks equities in emerging markets. A downturn in emerging markets can drive capital into Bitcoin as an alternative investment.

25.	Shanghai Composite Index

Tracks Chinese stock performance. Significant because of China’s historical influence on Bitcoin mining and trading.

26.	USD Index (DXY)

Measures the U.S. dollar’s strength against other major currencies. A stronger dollar often pressures Bitcoin downward, while a weaker dollar supports it.

27.	EUR to USD Exchange Rate

Reflects Euro-to-Dollar conversion. Can indirectly impact Bitcoin through currency strength dynamics and international investment flows.

**Bitcoin-Specific Outcome Variable**

28.	btc_close

Bitcoin’s closing price for a given period (e.g., daily, weekly). Used as the primary dependent variable for prediction models. All other features serve as explanatory variables to understand or forecast this value.

**Analysis Insight**

These features span multiple domains—sentiment, blockchain, technical, and macro—providing a holistic view of factors affecting Bitcoin prices. Each contributes uniquely to analyzing or predicting Bitcoin trends, allowing for a robust quantitative and qualitative evaluation.

```python
from statsmodels.tsa.seasonal import seasonal_decompose

# Decompose the bitcoin close price
result = seasonal_decompose(df['btc_close'], model='multiplicative', period=365)
fig = result.plot()
fig.set_size_inches((16, 9))
# Tight layout to realign things
fig.tight_layout()
plt.show()
```

1. raw data: big picture
2. trend: smooth version -> overall long-term trend
3. seasonal component: repeating short-term patterns
4. residuals: random noise after removing trend & seasonality (what we can't explain). If lack clear pattern, decomposition is successful.

```python
from statsmodels.tsa.stattools import adfuller

# Check stationary using ADF test
result = adfuller(df['btc_close'])

# Print results
print(f"ADF Statistic: {result[0]}")
print(f"p-value: {result[1]}")
print("Critical Values:")
for key, value in result[4].items():
    print(f"\t{key}: {value}")
```

adf stat not more negative than any of the critial values -> less stationary
p-value >= 0.05 -> non-stationary
- critical values: 	
  - 1% (-3.4321)
  - 5% (-2.8623)
  - 10% (-2.5672)
-> adf stat is < all critical values -> non-stationary

```python
# Apply first-order differencing
df['btc_close_diff'] = df['btc_close'].diff()

# Drop NaN values introduced by differencing
df = df.dropna(subset=['btc_close_diff'])

# Check the result
df[['btc_close', 'btc_close_diff']].head()
```

```python
# re-test for stationary
result_diff = adfuller(df['btc_close_diff'])
print(f"ADF Statistic: {result_diff[0]}")
print(f"p-value: {result_diff[1]}")
print("Critical Values:")
for key, value in result_diff[4].items():
    print(f"\t{key}: {value}")
```

now the p-value is < 0.05 -> stationary  
adf stat is < all critical values -> stationary  
-> stationary  
-> d=1 (from prior differencing): Indicates that the original series was differenced once to achieve stationarity.

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt

y = df['btc_close_diff']

# Plot ACF and PACF for btc_close_diff
plot_acf(y, lags=20)
plt.show()
plot_pacf(y, lags=20)
plt.show()
```

ACF plot: large spikes at 1 -> moving average q = 1  
PACF plot: large spikes at 1 -> autoregressive p = 1  
d = 1, p = 1, q = 1
